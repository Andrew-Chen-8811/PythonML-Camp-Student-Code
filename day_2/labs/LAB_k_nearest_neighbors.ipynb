```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import DecisionBoundaryDisplay
import pandas as pd
```

```python
df = pd.read_csv('data/knn_4_dataset.csv')  # Load the dataset
df.head(5)
```

```python
X = df[['feature_1', 'feature_2']].values
y = df['y'].values
```

```python
# Ensure X and y are defined
X = df[['feature_1', 'feature_2']].values
y = df['y'].values

plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KNN Dataset')
plt.show()
```

```python
# Instantiate and train K-NN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
```

```python
# Display the decision boundary and scatter the training points, similar to the lesson
plt.figure(figsize=(10, 7.5))
DecisionBoundaryDisplay.from_estimator(
    knn, X, response_method="predict", alpha=0.3
)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=20)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-NN Decision Boundary')
plt.show()
```
